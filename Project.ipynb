# Task 2
### Q2. The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task. 

```python
Project_df
```

```python
# Converting 'NaN' strings to actual NaN values:

Project_df["Cost"]=Project_df["Cost"].replace("nan",np.nan)
```

```python
# Convert the column to numeric.

Project_df["Cost"]=pd.to_numeric(Project_df["Cost"])
```

```python
# Calculating Average while ignoring NaN values (Sum/12)

a1= Project_df["Cost"].mean()
print("Calculated Average =",a1)
```

### Above we calculated average of cost column, but we need running average which is calculated below.
#### why running avg? 
   * Real-time estimation: The missing value is replaced based on previously observed values.
* Prevents bias: Unlike a static mean (df['Cost'].mean()), this method adapts dynamically.

```python
# Convert 'Cost' to numeric (in case of string NaN values)
Project_df['Cost'] = pd.to_numeric(Project_df['Cost'], errors='coerce')

# Running sum and count to compute the running average
running_sum = 0
running_count = 0

# Iterate through the DataFrame using a for loop
for i in range(len(Project_df)):   # len(Project_df) gives the total number of rows in the DataFrame.    #iterates over each row index (i)
    if np.isnan(Project_df.loc[i, 'Cost']):  # Check if value is NaN, True/False, This retrieves the value of the 'Cost' column for the row indexed by i
        if running_count > 0:
            Project_df.loc[i, 'Cost'] = running_sum / running_count  # Replace NaN with running average
    else:
        running_sum += Project_df.loc[i, 'Cost']  # Update running sum
        running_count += 1  # Update running count


# Converting the column to int so as to avoid scientific notation of cost column.
Project_df['Cost'] = Project_df['Cost'].astype(int)

# Display the updated DataFrame
Project_df
```

#### alternatively using Pandas library which is more efficient way

```python
Project_df['Cost'] = Project_df['Cost'].fillna(Project_df['Cost'].expanding().mean())

print("DataFrame after replacing NaN using Pandas:")
print(Project_df)
```

# Task 3
### Q3. Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column. 

```python
Employee_df
```

```python
# Splitting 'Name' column into 'First Name' and 'Last Name'
Employee_df[['First Name','Last Name']] = Employee_df['Name'].str.split(' ', n=1, expand=True)   
# split at first space for once(n=1)
# n=1 ensures only two parts (First & Last Name) are created.
# expand=True converts the result into separate columns.

# Dropping the original 'Name' column
Employee_df.drop(columns=['Name'], inplace=True)

# Reordering columns to place 'First Name' and 'Last Name' after 'ID'
desired_order = ['ID', 'First Name', 'Last Name'] + [col for col in Employee_df.columns if col not in ['ID', 'First Name', 'Last Name']]
Employee_df = Employee_df[desired_order]

Employee_df
```

# Task 8
### Q8 Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

#### Here for promotion we have to decrease the designation level by 1 as designation scale is in descending order (1-highest)

```python
# Converting Age column to integer type so as to avoid scietific notation.
Final['Age'] = Final['Age'].astype(int)

# Defining function for promoting designation level by 1 for employees whose Age > 29
def promote_designation(row):
    if row['Age'] > 29:
        return row['Designation Level'] - 1
    return row['Designation Level']

# Applying function to "Designation Level" column.
Final['Designation Level'] = Final.apply(promote_designation, axis=1)
```
